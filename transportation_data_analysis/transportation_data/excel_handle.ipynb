```python
dir_str = r'E:/xinjiang/data/data_handle_211101/新疆交投观测站数据/新疆交投观测站数据/重点节假日/2020年5月日交通量'
```

```python
import os
file_name = os.listdir(dir_str)
```

```python
file_name
```

```python
import pandas as pd
import numpy as np
import warnings
import datetime
warnings.filterwarnings('ignore')
```

```python
data_above = pd.DataFrame()
for i in range(len(file_name)):
    excecl = pd.read_excel(dir_str+'/'+file_name[i])
    sheet_sum = excecl[['连续观测站交通量月度统计报表','Unnamed: 2',"Unnamed: 3","Unnamed: 5","Unnamed: 6","Unnamed: 7","Unnamed: 9","Unnamed: 10"]].iloc[10:42]
    index_list = []
    for j in range(len(sheet_sum['连续观测站交通量月度统计报表'])):
        if type(sheet_sum['连续观测站交通量月度统计报表'].iloc[j]) != datetime.datetime:
            index_list.append(sheet_sum.index[j])
    sheet_sum.drop([k for k in index_list],axis=0,inplace=True)
    sheet_sum.columns = ['日期','货小','货中','货大','货特大','集装箱','客小','客大']
    sheet_sum['标签'] = file_name[i]
    data_above = pd.concat([data_above,sheet_sum],ignore_index=True)
```

```python
data_above['货车'] = data_above.iloc[:,1:6].sum(axis=1)
```

```python
data_above['客车'] = data_above.iloc[:,6:8].sum(axis=1)
```

```python
data_above
```

```python
data_above_classify = data_above.drop(["货小","货中","货大","货特大","集装箱","客小","客大"],axis=1)
data_above_classify
data_melt = data_above_classify.melt(id_vars=['日期','标签'])
```
